```python
employee = employee.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis = 1)
```

```python
employee.columns
```

```python
missing_value_presentage = round(employee.isnull().sum()*100/len(employee),2).reset_index()
missing_value_presentage.columns = ['column_name','missing_value_presentage']
missing_value_presentage = missing_value_presentage.sort_values('missing_value_presentage',ascending =False)
missing_value_presentage
```

```python
# replacing the attrition output by integer constants
employee.loc[employee['Attrition']=='No','Attrition'] = 0
employee.loc[employee['Attrition']=='Yes','Attrition'] = 1
```

```python
# take out our numeric columns
employee_num = employee[employee.select_dtypes(include=[np.number]).columns.tolist()]
```

```python
employee_num.head()
```

```python
employee_dummies = pd.get_dummies(employee[employee.select_dtypes(include=['object']).columns.tolist()])
```

```python
employee_dummies.head()
```

```python
employee_combined = pd.concat([employee_num,employee_dummies],axis = 1)
```

```python
employee_combined.head()
```

```python
# dividing the dataset into train and test dataset
from sklearn.model_selection import train_test_split
```

```python
train_x = employee_combined.drop(['Attrition'],axis = 1)
```

```python
train_y = employee_combined['Attrition']
```

```python
X_train,X_test,y_train,y_test = train_test_split(train_x,train_y,test_size = 0.3,random_state = 20)
```

```python
# import the tree classifier model
from sklearn import tree
```

```python
dt = tree.DecisionTreeClassifier()
```

```python
# train the model using trainig data
dt.fit(X_train,y_train)
```

```python
# predict the responce for the test dataset
y_pred = dt.predict(X_test)
```

```python
# import scikitlearn metrics for calculate accuracy
from sklearn import metrics
```

```python
# model accuracy,how many times the model is correct
print('Accuracy : ',metrics.accuracy_score(y_test,y_pred))
```

```python
print('Precision : ',metrics.precision_score(y_test,y_pred))
```

```python
print('Recall : ',metrics.recall_score(y_test,y_pred))
```
